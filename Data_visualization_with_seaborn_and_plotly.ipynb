# Overview
Data visualization is essential for understanding patterns in data. This class note covers statistical plotting using Seaborn and Plotly, focusing on relational, distributional, and categorical plots
to derive insights, and converting Seaborn plots to Plotly for interactive visualizations.



# Objectives

• Create various statistical plots using Seaborn.

• Understand insights derived from statistical plots.

• Convert Seaborn plots to equivalent Plotly visualizations.

#Key Concepts
1.**Relational Plots:** Visualize relationships between variables.

• Scatter Plot: Shows correlation between two variables (e.g., total bill vs. tip).

• Line Plot: Displays trends over a continuous variable (e.g., time vs. signal).

2**.Distributional Plots**: Analyze variable distributions.

• Univariate: Distribution of one variable (e.g., histogram of flipper length).

• Bivariate: Distribution of two variables (e.g., bill length vs. depth).

• Multivariate: Pair plots for multiple variables.

3.**Categorical Plots:** Explore relationships involving categorical data.

• Strip/Swarm: Individual data points by category.

• Box/Violin: Distribution summaries by category.

• Count: Frequency of categories.

4.**Plotly:** Interactive visualizations, equivalent to Seaborn plots.

```python
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
"""Congigure seaborn setup"""
sns.set(
    style="darkgrid",
    palette="flare",
    font="sans-serif",
    font_scale=1,
)
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
tips["time"].value_counts()
```

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
penguins = sns.load_dataset("penguins")
penguins.head()
```

### **Relational plot(relplot)**

```python
"""scatter plot (Seaborn)"""
plot = sns.relplot(
  data=tips,
  x="total_bill",
  y="tip",
  kind="scatter",
  height=4,


)
plot.set(
      xlabel="Total Bill",
      ylabel="Tip",
      title="Total Bill vs Tip"
)
```

#hue

```python
plot = sns.relplot(
  data=tips,
  x="total_bill",
  y="tip",
  kind="scatter",
  height=4,
  hue="day",

)
plot.set(
      xlabel="Total Bill",
      ylabel="Tip",
      title="Total Bill vs Tip"
)
```

#col

```python
plot = sns.relplot(
  data=tips,
  x="total_bill",
  y="tip",
  kind="scatter",
  height=4,
  hue="day",
  col="day"

)
plot.set(
      xlabel="Total Bill",
      ylabel="Tip",
      title="Total Bill vs Tip"
)
```

#row

```python
plot = sns.relplot(
  data=tips,
  x="total_bill",
  y="tip",
  kind="scatter",
  height=4,
  hue="sex",
  row="day"

)
plot.set(
      xlabel="Total Bill",
      ylabel="Tip",
      title="Total Bill vs Tip"
)
```

```python
plot = sns.relplot(
  data=tips,
  x="total_bill",
  y="tip",
  kind="scatter",
  height=4,
  hue="sex",
  col="day"

)
plot.set(
      xlabel="Total Bill",
      ylabel="Tip",
      title="Total Bill vs Tip"
)
```
